```python
def roi(cost, revenue):
  assert cost != 0, "Cost cannot be zero"
  assert isinstance(cost, (int, float)), "Cost must be an int or float"
  assert isinstance(revenue, (int, float)), "Revenue must be an int or float"
  return (revenue - cost) / cost
roi(5, 10)
```

```python
def loan_payment(principal, annual_rate, months):
  assert isinstance(principal, (int, float)), "Principal must be an int or float"
  assert isinstance(annual_rate, (int, float)), "Annual rate must be an int or float"
  assert isinstance(months, int), "Months must be an int"
  if (principal <= 0) or (months<= 0) or (annual_rate < 0):
    raise ValueError("Invalid input")
  try:
    r = annual_rate / 12
    if r == 0:
      pmt = principal / months
    else:
      pmt = r * principal / (1 - (1 + r) ** (-months))
  except Exception as e:
    print(f"Error: {e}")
  return pmt
loan_payment(10000, 0.05, 12)
```

```python
def getValues(n):
  assert n > 0, "n must be a positive integer"
  assert isinstance(n, int), "Input variable 'n' must be an integer"
  items = []

  for i in range(n):
    items.append(float(input("add a value to the list: ")))
  return items

def calcAverage(alist):
  assert len(alist) > 2, "List must have at least 3 items"
  assert type(alist) == list, "Input variable 'alist' must be a list"
  try:
    return sum(alist[2:]) / len(alist[2:])
  except Exception as e:
    print(f"Error: {e}")
    raise ValueError("The length of list 'alist' must be greater than 2")

def main():
  n = int(input("How many weights will you enter? "))
  alist = getValues(n)
  avg = calcAverage(alist)
  print(avg)

main()
```

```python
class Greeting:
  def __init__(self, first_name, last_name):
    self.first_name = first_name
    self.last_name = last_name

  def display(self, isFormal):
    if isFormal:
      print(f"{self.last_name}, {self.first_name}")
    else:
      print(self.first_name)

  def __str__(self):
    return f"{self.first_name} {self.last_name}"
```

```python
user = Greeting("John", "Doe")
user.display(True)
print(user)
```

```python
class Student:
  def __init__(self, name, number):
    assert isinstance(name, str)
    assert isinstance(number, int)
    self.__name = name
    self.number = number
    self.__courses = []

  def get_name(self):
    return self.__name

  def get_courses(self):
    return self.__courses

  def set_name(self, newName):
    self.__name = newName

  def __len__(self):
    return len(self.__courses)

  def enroll(self, newCourse):
    if newCourse not in self.__courses:
      self.__courses.append(newCourse)
      print(f"Student {self.__name} enrolled in {newCourse}")
    else:
      print(f"Student {self.__name} is already enrolled in {newCourse}")
```

```python
a = Student("John Doe", 12345)
a.enroll("Science")
print(a.get_courses())
print(a.__len__())
a.enroll("Science")
print(a.get_courses())
print(a.__len__())
```

```python
import math

class Point:
  def __init__(self, x, y):
    self.x = x
    self.y = y

  def get_distance(self, other_point):
    distance = math.sqrt(((self.x - other_point.x) ** 2) + ((self.y - other_point.y) ** 2))
    return f"Distance between ({self.x},{self.y}) and ({other_point.x},{other_point.y}) is {distance}"

class Rectangle:
  def __init__(self, top_left, top_right, bottom_right, bottom_left):
    self.top_left = top_left
    self.top_right = top_right
    self.bottom_right = bottom_right
    self.bottom_left = bottom_left

  def width(self):
    return math.sqrt(((self.top_right.x - self.top_left.x) ** 2) + ((self.top_right.y - self.top_left.y) ** 2))

  def height(self):
    return math.sqrt(((self.bottom_left.x - self.top_left.x) ** 2) + ((self.bottom_left.y - self.top_left.y) ** 2))

  def area(self):
    return self.width() * self.height()
```

```python
a = Point(0, 0)
b = Point(3, 4)
a.get_distance(b)
```

```python
r = Rectangle(
  top_left = Point(0, 4),
  top_right = Point(5, 4),
  bottom_right = Point(5, 0),
  bottom_left = Point(0, 0)
)

print(r.width())   # 5.0
print(r.height())  # 4.0
print(r.area())    # 20.0
```

```python
class Employee:
  def __init__(self, name, wage, hours):
    self.name = name
    self.__wage = wage
    self.hours = hours

  def calcPay(self):
    return self.__wage * self.hours

  def __str__(self):
    return(f"Name: {self.name}, Hourly Wage: {self.__wage}")

  def get_wage(self):
    return self.__wage

  def set_wage(self, new_wage):
    assert new_wage > 0, "Wage must be greater than 0"
    self.__wage = new_wage

class Manager(Employee):
  def __init__(self, name, wage, hours, bonus):
    super().__init__(name, wage, hours)
    self.bonus = bonus #assume it is given as float from 0 -> 1 (e.g. 20% bonus is 0.2)

  def calcPay(self):
    return (1+self.bonus) * self.get_wage() * self.hours # Optional way  is (1 + self.bonus) * super().calcPay()

employee = Employee("Bob", 50, 10)
manager = Manager("Alice", 100, 15, 0.2)
print(employee)
print(manager)
print()
print(f"{employee.name}'s weekly salary is ${employee.calcPay()}.")
print(f"{manager.name}'s weekly salary is ${manager.calcPay()} that includes a {manager.bonus * 100}% bonus.")
```
